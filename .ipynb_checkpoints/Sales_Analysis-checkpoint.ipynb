```python
ad = "Datasets/Ad3.csv"
goods = "Datasets/cost_of_goods.csv"
transactions="Datasets/transactions_main.csv"

ad_data =pd.read_csv(ad, index_col=False)
goods_raw =pd.read_csv(goods)
transactions_raw =pd.read_csv(transactions)

ad_df = pd.DataFrame(ad_data)
goods_df = pd.DataFrame(goods_raw)
trans_df = pd.DataFrame(transactions_raw)

df =pd.merge(trans_df, goods_df, how='outer', on='sku', left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
```

```python
df
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.loc[:, 'date/time'].replace(regex=True,inplace=True,to_replace="PST", value="")
df.loc[:, 'date/time'].replace(regex=True,inplace=True,to_replace="PDT", value="")
df['date/time']=pd.to_datetime(df['date/time'])
df.infer_objects();
```

```python
def trans(x):
    x['transaction revenue']= x['total']-x['est_cost_unit']
    x['sale price']= x['product sales']/x['quantity']
    x['Cost of Goods']=x['quantity']*x['est_cost_unit']
    x['Whole Date'] =pd.to_datetime(df['date/time']).apply(lambda y: y.date())
```

```python
trans(df)
trans_df=df.drop(['settlement id','order id','marketplace','fulfillment'],axis=1)
trans_df.sort_values('date/time').reset_index().head(5)
trans_df['sale price'] =trans_df['sale price'].round(2)
```

```python
adclean = ad_data
adclean['Date'] = pd.to_datetime(adclean['Date'], errors='coerce')
adclean = adclean.dropna(subset=['Date'])
adclean['Date'] = adclean['Date'].apply(lambda x: x.date())

def ad_cols(x):
    x['Whole Date'] = pd.to_datetime(x['Date']).apply(lambda y: y.date())
    return x
```

```python
ad_cols(adclean)
short_ad=adclean.drop(['Currency'],axis=1)
short_ad['Spend']=short_ad['Spend'].replace('[\$,]', '', regex=True).astype(float)
short_ad.head(5)
```
